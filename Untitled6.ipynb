#Gender

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data=df,x="Gender")
plt.show()
```

```python
ax=sns.countplot(data=df,x="Gender")
for bars in ax.containers: ax.bar_label(bars)
```

```python
sales_gen=df.groupby (["Gender"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_gen
```

```python
sns.barplot(data=sales_gen,x="Gender",y="Amount")
plt.show()
```

From Above  Graph we can see that most of the customers are females and even the purchasing power of females are greater than men

# State

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.countplot(data=df.head(10),x="State",hue="Orders")
plt.show()
```

```python
# Total numbers 0f orders from top 10 states

sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sales_state
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=sales_state,x="State",y="Orders")
plt.show()
```

```python
# total amount sales from top 10 states

sales_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sales_state
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=sales_state,x="State",y="Amount")
plt.show()
```

#from above graph we can see that unexpectedly most of the orders are from uttar pradesh,Maharastra and karnataka
respectievely  but total sales/amount from up,karnataka then maharastra

```python
plt.figure(figsize=(20,5))
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data=df,x="Product_Category")
plt.show()
```
